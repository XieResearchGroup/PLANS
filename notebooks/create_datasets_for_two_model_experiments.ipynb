# Label all negative data as 0 and all other data (at least one class is positive) as 1

```python
import pandas as pd
import numpy as np
```

### Load data

```python
cyp_data_path = "../data/fromraw_cid_inchi_smiles_fp_labels_onehots.csv"
cyp_df = pd.read_csv(cyp_data_path)
cyp_df.head()
```

### Create labels

```python
def two_class_label(label):
    if "_" in label:
        return np.nan
    elif label == "00000":
        return "0"
    else:
        return "1"
```

```python
cyp_df["two_class_label"] = cyp_df["Label"].map(two_class_label)
```
