```python
import os
import json

import pandas as pd
```

```python
cyp450_cvs_path = os.path.join("..", "data", "fromraw_cid_inchi_smiles_fp_labels_onehots.csv")
output_path = os.path.join("..", "data", "cyp450_smiles_ECFP_labels.json")
```

```python
oridata = pd.read_csv(cyp450_cvs_path)
```

```python
oridata.head()
```

```python
jsondata = list()
for rid, row in oridata.iterrows():
    jsondata.append([row["isomeric_SMILES"], list(map(int, list(row["ECFP"]))), row["Label"]])
```

```python
with open(output_path, "w") as f:
    json.dump(jsondata, f)
```
